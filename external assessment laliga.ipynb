#### 2. Print all the teams which have started playing between 1930-1980. (5 points)

```python
b= laliga[['Team','Debut']]
```

```python
# List of years
year=1929
y=[]
for i in range(51):
    year +=  1
    y.append(year)
```

```python
# Calculation to find the teams which have started playing
inclusiveCount = []
exclusiveCount = []
for i in range(len(b)):
    if pd.to_numeric(b.iloc[:,-1][i][:4]) in y:
        inclusiveCount.append(b.iloc[i])
    if pd.to_numeric(b.iloc[:,-1][i][:4]) in y[1:-1]:
        exclusiveCount.append(b.iloc[i])        
print('No. of teams which have started playing between 1930-1980. (inclusive) : ', len(inclusiveCount))
print('No. of teams which have started playing between 1930-1980. (exclusive) : ', len(exclusiveCount))
```

#### 3. Print the list of teams which came Top 5 in terms of points (2.5 points)

```python
# Taking first 5 points
top5Points = pd.to_numeric(laliga['Points']).nlargest(5).astype('str')
```

```python
# Printing the list of team which are matching the first 5 points
top5Team = []
for i in range(len(laliga)):    
    for j in range(len(top5Points)): 
        if (laliga['Points'][i] == top5Points[j] ):
            top5Team.append(laliga['Team'][i])
        
pd.DataFrame({'Top 5 Team' : top5Team})
```

```python
print(pd.DataFrame({'Top 5 Team' : top5Team}))
```

#### 4. Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference. (5 points)
Goal_diff_count = GoalsFor - GoalsAgainst
If there are any numerical error, replace it with 0%

```python
# function which returns the Team name and Goal difference

def Goal_diff_count():
    listGoals = pd.DataFrame()
    goalDiff =[]
    for i in range(len(laliga)):
        goalDiff.append(pd.to_numeric(laliga['GoalsFor'][i])-pd.to_numeric(laliga['GoalsAgainst'][i]))       
    listGoals['Team'] = laliga['Team']
    listGoals['GoalDiff'] = goalDiff    
    return listGoals
```

```python
# Maximum goal difference
Goal_diff_count().sort_values('GoalDiff', ascending = False).head(5)
```

```python
# Miniumm Goal Diference
Goal_diff_count().sort_values('GoalDiff', ascending = True).head(5)
```

#### 5. Create a new column with name “Winning Percent” and append it to the data set (5 points)
Percentage of Winning = (GamesWon / GamesPlayed)*100

If there are any numerical error, replace it with 0%

Print the top 5 teams which has the highest Winning percentage

```python
# Created new column WinningPercent and appended it to the dataset
WP = []
W = pd.to_numeric(laliga['GamesWon'])
P = pd.to_numeric(laliga['GamesPlayed'])

for i in range(len(W)):
    WP.append(((W[i]/P[i])*100.00))
    
laliga['WinningPercent'] = WP

laliga['WinningPercent'] = laliga[['WinningPercent']].fillna('0').astype('int64').astype('str')+'%'


laliga[['Team','WinningPercent']]
```

```python
# Top 5 teams with high WinningPercent

print(laliga.sort_values(['WinningPercent'], ascending=False)[['Team','WinningPercent']].iloc[:5])
```
