# Q8 Split the pima_df into training and test set in the ratio of 70:30 (Training:Test).


```python
X = mydata.iloc[:,:-1]
Y = mydata.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size =0.3)

print(x_train.shape)
print(x_test.shape)
```

# Q9 Create the decision tree model using “entropy” method of reducing the entropy and fit it to training data.</b>



```python
dt_model = DecisionTreeClassifier(criterion = 'entropy',max_depth =2)
#dt_model = DecisionTreeClassifier()
dt_model.fit(x_train, y_train)
```

# Q10 Print the accuracy of the model & print the confusion matrix 

```python
dt_model.score(x_test , y_test)
```

```python
dt_model.score(x_train , y_train)
```

```python
from sklearn.metrics import confusion_matrix 
y_predict = dt_model.predict(x_test)
```

```python
confusion_matrix(y_test, y_predict)
```

Print the feature importance of the decision model - Optional

# Q12 Apply the Random forest model and print the accuracy of Random forest Model

```python
rfcl = RandomForestClassifier(n_estimators = 85)
rfcl = rfcl.fit(x_train, y_train)
```

```python
rfcl.score(x_test , y_test)
```

```python
y_predict = rfcl.predict(x_test)
confusion_matrix(y_test, y_predict)
```

# Q13 Apply Adaboost Ensemble Algorithm for the same data and print the accuracy.

```python
abcl = AdaBoostClassifier()
abcl = abcl.fit(x_train, y_train)

```

```python
test_pred = abcl.predict(x_test)
abcl.score(x_test , y_test)
```

# Q14 - Apply Bagging Classifier Algorithm and print the accuracy.

```python
from sklearn.ensemble import BaggingClassifier
bgcl = BaggingClassifier(n_estimators=100,  oob_score=True , random_state=1)

bgcl = bgcl.fit(X, Y)

print(bgcl.oob_score_)
print(bgcl.score(x_test, y_test))
print(bgcl.score(x_train, y_train))
```
