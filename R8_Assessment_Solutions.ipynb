# Text classification using TF-IDF

### 1. Load the dataset from sklearn.datasets

```python
from sklearn.datasets import fetch_20newsgroups
```

```python
categories = ['alt.atheism', 'soc.religion.christian', 'comp.graphics', 'sci.med']
```

### 2. Training data

```python
twenty_train = fetch_20newsgroups(subset='train', categories=categories, shuffle=True, random_state=42)
```

### 3. Test data

```python
twenty_test = fetch_20newsgroups(subset='test', categories=categories, shuffle=True, random_state=42)
```

###  a.  You can access the values for the target variable using .target attribute 
###  b. You can access the name of the class in the target variable with .target_names


```python
twenty_train.target
```

```python
twenty_train.target_names
```

```python
twenty_train.data[0:5]
```

### 4.  Now with dependent and independent data available for both train and test datasets, using TfidfVectorizer fit and transform the training data and test data and get the tfidf features for both

```python
from sklearn.feature_extraction.text import TfidfVectorizer 
tfidf_vectorizer=TfidfVectorizer(stop_words = 'english')
X_train_tfidf=tfidf_vectorizer.fit_transform(twenty_train.data)
X_test_tfidf=tfidf_vectorizer.transform(twenty_test.data)

print(X_train_tfidf.shape)
print(X_test_tfidf.shape)
```

```python
print(tfidf_vectorizer.get_feature_names())
```

### 5. Use logisticRegression with tfidf features as input and targets as output and train the model and report the train and test accuracy score

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
```

```python
logreg.fit(X_train_tfidf,twenty_train.target)
```

```python
y_pred = logreg.predict(X_train_tfidf)
y_pred.shape
```

```python
y_pred = logreg.predict(X_test_tfidf)
y_pred.shape
```

```python
print('Accuracy of logistic regression classifier on train set: {:.2f}'.format(logreg.score(X_train_tfidf, twenty_train.target)))
print('Accuracy of logistic regression classifier on test set: {:.2f}'.format(logreg.score(X_test_tfidf, twenty_test.target)))
```
